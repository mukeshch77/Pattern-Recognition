```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

## Linear Regression Curve fitting 

### Generating synthetic data (linear relationship)

```python
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Random independent variable between 0 and 10
y = 2 * X + 1 + np.random.randn(100, 1)  # y = 2 * X + 1 + noise
```

### Fitting a linear regression model


```python
linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
y_pred = linear_regressor.predict(X)
```

### Visualizing the linear regression line

```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression')
plt.legend()
plt.show()
```

### Compute MSE and MAE

```python
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)

print(f"Linear Regression - MSE: {mse:.3f}, MAE: {mae:.3f}")
```

## Non-Linear Regression Curve fitting 

```python
from sklearn.preprocessing import PolynomialFeatures
```

### Generating synthetic data (quadratic relationship)


```python
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Independent variable
y = 0.5 * X**2 - 3 * X + 2 + np.random.randn(100, 1)  # y = 0.5 * X^2 - 3 * X + 2 + noise
```

### Transforming X to include polynomial features (degree=2 for quadratic)


```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

### Fit a polynomial regression model


```python
poly_regressor = LinearRegression()
poly_regressor.fit(X_poly, y)
y_poly_pred = poly_regressor.predict(X_poly)
```

### Visualizing the polynomial regression curve


```python
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(np.sort(X, axis=0), poly_regressor.predict(poly.transform(np.sort(X, axis=0).reshape(-1, 1))), color='red', label='Fitted Curve')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Non-Linear Polynomial Regression')
plt.legend()
plt.show()
```

### Compute MSE and MAE for the polynomial regression


```python
mse_poly = mean_squared_error(y, y_poly_pred)
mae_poly = mean_absolute_error(y, y_poly_pred)

print(f"Polynomial Regression - MSE: {mse_poly:.3f}, MAE: {mae_poly:.3f}")
```

### Manually calculate MSE and MAE


```python
# Linear Regression
def calculate_mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def calculate_mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))


mse_manual = calculate_mse(y, y_pred)
mae_manual = calculate_mae(y, y_pred)

print(f"Manual Calculation - Linear Regression - MSE: {mse_manual:.3f}, MAE: {mae_manual:.3f}")
```

### Display results

```python
print(f"Linear Regression - MSE: {mse:.3f}, MAE: {mae:.3f}")
print(f"Polynomial Regression - MSE: {mse_poly:.3f}, MAE: {mae_poly:.3f}")
print(f"Manual Calculation - Linear Regression - MSE: {mse_manual:.3f}, MAE: {mae_manual:.3f}")
```
